```python
import numpy as np 
import pandas as pd 
import cv2
import matplotlib.pyplot as plt 
import tensorflow as tf 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Flatten, Dropout, Activation, Conv2D, MaxPooling2D 

import os 
print(os.getcwd())
```

```python
trainlist = pd.read_csv("D:\\shopee-product-detection-student\\train.csv")
```

```python
trainlist
```

```python
trainlist.category = trainlist.category.astype("category")
trainlist.category = trainlist.category.apply(lambda c: str(c).zfill(2))
```

```python
x = []
x = trainlist.filename.to_numpy()
len(x)
```

```python
x = []
y = []
z = []

path = "D:\\shopee-product-detection-student\\train"
x = trainlist.filename.to_numpy()
for j in trainlist.category: 
    y.append(j)
#y = trainlist.category.to_numpy()

for i in range(len(x)): 
    img_array = cv2.imread(os.path.join(path, y[i],x[i]), cv2.IMREAD_COLOR)
    new_img_array = cv2.resize(img_array, dsize = (80,80))
    z.append(new_img_array)
   
z = np.array(z).reshape(-1, 80, 80, 1)

    
 

```

```python
y = np.array(y).astype(np.float32)
type(y[0])
```

```python
model = Sequential()
#add a densely-connected layer with 64 units to the model: 
model.add(Conv2D(64,(3,3), activation = 'relu', input_shape = z.shape[1:]))
model.add(MaxPooling2D(pool_size = (2,2)))
#add another layer
model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Flatten())

#add a softmax layer with 10 output units 
model.add(Dense(64, activation ='relu'))
model.add(Dense(42, activation = 'softmax'))
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(z, y, epochs = 10, batch_size = 320, validation_split = 0.2)
```
